```python
# Feature engineering few parameters
df['Spending']=df['MntWines']+df['MntFruits']+df['MntMeatProducts']+df['MntFishProducts']+df['MntSweetProducts']+df['MntGoldProds']
```

```python
df=df.rename(columns={'NumWebPurchases': "Web",'NumCatalogPurchases':'Catalog','NumStorePurchases':'Store'})
```

```python
df['Marital_Status']=df['Marital_Status'].replace({'Divorced':'Alone','Single':'Alone','Married':'In couple','Together':'In couple','Absurd':'Alone','Widow':'Alone','YOLO':'Alone'})
df['Education']=df['Education'].replace({'Basic':'Undergraduate','2n Cycle':'Undergraduate','Graduation':'Postgraduate','Master':'Postgraduate','PhD':'Postgraduate'})
```

```python
df=df.rename(columns={'MntWines': "Wines",'MntFruits':'Fruits','MntMeatProducts':'Meat','MntFishProducts':'Fish','MntSweetProducts':'Sweets','MntGoldProds':'Gold'})
```

```python
df['Children']=df['Kidhome']+df['Teenhome']
df['Has_child'] = np.where(df.Children> 0, 'Has child', 'No child')
df
```

```python
# creating age variable
df['Age']=2014-df['Year_Birth']
df
```

```python
# calculating the loyalty of customer in terms of seniority i.e., how long the customer have been a part of the firm
from datetime import date
last_date = date(2014,10, 4)
df['Seniority']=pd.to_datetime(df['Dt_Customer'], dayfirst=True,format = '%Y-%m-%d')
df['Seniority'] = pd.to_numeric(df['Seniority'].dt.date.apply(lambda x: (last_date - x)).dt.days, downcast='integer')/30
```

```python
df
```

```python
dfn = df[['Age','Education','Marital_Status','Income','Spending','Seniority','Has_child','Children','Wines','Fruits','Meat','Fish','Sweets','Gold']]
```

```python
dfn.head()
```

```python
# We can divide customers into 4 segements i.e, 
# loyal_expenders - rich and high spends
# new_low - new with low income but low spends
# new_high - new with high income and high spends
# loyal_unexpenders - below average income and low spends
```

```python
import numpy as np

# Define medians
income_median = dfn['Income'].median()
spending_median = dfn['Spending'].median()
seniority_median = dfn['Seniority'].median()

# Define conditions
conditions = [
    (dfn['Seniority'] >= seniority_median) & 
    (dfn['Income'] > income_median) & 
    (dfn['Spending'] > spending_median),
    
    (dfn['Seniority'] < seniority_median) & 
    (dfn['Income'] > income_median) & 
    (dfn['Spending'] > spending_median),
    
    (dfn['Seniority'] >= seniority_median),
    
    (dfn['Seniority'] < seniority_median)
]

# Define corresponding labels
choices = [
    'loyal_expenders',
    'new_high',
    'loyal_unexpenders',
    'new_low'
]

# Apply segmentation
dfn['Segment'] = np.select(conditions, choices, default='unknown')
```

```python
# removing outliers
# Calculate Q1, Q3, and IQR for Income and Spending
Q1_income = dfn['Income'].quantile(0.25)
Q3_income = dfn['Income'].quantile(0.75)
IQR_income = Q3_income - Q1_income

Q1_spending = dfn['Spending'].quantile(0.25)
Q3_spending = dfn['Spending'].quantile(0.75)
IQR_spending = Q3_spending - Q1_spending

# Define outlier bounds for Income and Spending
lower_bound_income = Q1_income - 1.5 * IQR_income
upper_bound_income = Q3_income + 1.5 * IQR_income

lower_bound_spending = Q1_spending - 1.5 * IQR_spending
upper_bound_spending = Q3_spending + 1.5 * IQR_spending

# Identify outliers in Income and Spending
income_outliers = dfn[(dfn['Income'] < lower_bound_income) | (dfn['Income'] > upper_bound_income)]
spending_outliers = dfn[(dfn['Spending'] < lower_bound_spending) | (dfn['Spending'] > upper_bound_spending)]

dfn = dfn[(dfn['Income'] >= lower_bound_income) & (dfn['Income'] <= upper_bound_income) &
                   (dfn['Spending'] >= lower_bound_spending) & (dfn['Spending'] <= upper_bound_spending)]
```
